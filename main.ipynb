```python
import numpy as np
import pandas as  pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("netflix_titles.csv")

df = pd.DataFrame(data)
print(df)
```

```python
df.describe()
```

```python
df['country'] = df['country'].fillna("Unknown")

df['date_added'] = df['date_added'].fillna(df['date_added'].mode()[0])
df['rating'] = df['rating'].fillna(df['rating'].mode()[0])
df['duration'] = df['duration'].fillna(df['duration'].mode()[0])
```

```python
df['type'].value_counts()
release_year =df.groupby(['release_year', 'type']).size().unstack()
print(release_year)



```

```python
df.groupby(['release_year', 'type']).size().unstack().plot(kind='bar', figsize=(12,6))
```

* frame â†’ your DataFrame

* id_vars â†’ columns to keep

* value_vars â†’ columns to melt

* var_name â†’ name for the new â€œcolumn namesâ€ column

* value_name â†’ name for the new â€œvaluesâ€ column

```python


release_year_reset = release_year.reset_index()


release_year_long = release_year_reset.melt(
    id_vars='release_year', 
    value_vars=['Movie', 'TV Show'], 
    var_name='type', 
    value_name='count'
)


release_year_long = release_year_long.dropna()

data_90_to_21 = release_year_long[release_year_long["release_year"] > 1990]



plt.figure(figsize=(15,6))
sns.barplot(data=data_90_to_21, x='release_year', y='count', hue='type')
plt.xticks(rotation=90)  
plt.title('Number of Movies and TV Shows from 1991 to 2021')
plt.show()
```

```python
country =df.groupby(['country', 'type']).size().unstack().reset_index().dropna()
print(country)

country_long = country.melt(
    id_vars='country', 
    value_vars=[ 'Movie',"TV Show"], 
    var_name='type', 
    value_name='count'
)
print(country_long)
```

```python

avg_count = country_long['count'].mean()


above_avg_count = country_long[country_long['count'] > avg_count]


```

```python
plt.figure(figsize=(15,6))
sns.barplot(data=above_avg_count, x='country', y='count', hue='type')
plt.xticks(rotation=90)  
plt.title('Number of Movies and TV Shows by country')
plt.ylabel("Number of shows produce")
plt.xlabel("Country")
plt.show()
```

```python
df["listed_in"]
```

```python
genres = ["Action & Adventure","Anime Features","Anime Series","British TV Shows","Children & Family Movies",
          "Classic & Cult TV","Classic Movies","Comedies","Crime TV Shows","Cult Movies","Documentaries",
          "Docuseries","Dramas","Horror Movies","Independent Movies","International Movies",
          "International TV Shows","Kid's TV","Reality TV","Romantic TV Shows","TV Action & Adventure",
          "TV Comedies","TV Dramas","TV Horrors"]

df['main_genre'] = df['listed_in'].apply(
    lambda x: next((g for g in genres if g.lower() in x.lower()), 'other')
)

print(df['main_genre'].value_counts())
```

```python
listed =df.groupby(['type', 'main_genre']).size().unstack().reset_index()
print(listed)
```

```python
df_melt = listed.melt(id_vars='type', var_name='genre', value_name='count')
df_melt = df_melt.dropna()
print(df_melt)
```

```python
plt.figure(figsize=(14,6))
sns.barplot(x='genre', y='count', hue='type', data=df_melt)
plt.title('Comparison of Movies vs TV Shows by Genre')
plt.xticks(rotation=75)
plt.xlabel('Genre')
plt.ylabel('Number of Titles')
plt.tight_layout()
plt.show()
```

```python
director_list =df["director"].value_counts().reset_index(name="count")


d_avg = director_list.head(10)
print(d_avg)
```

```python

plt.figure(figsize=(14,6))
sns.barplot(data=d_avg, x='director', y='count', palette='mako')
plt.title('Top 10 Most Frequent Directors on Netflix')
plt.xlabel('Director')
plt.ylabel('Number of Shows/Movies')
plt.xticks(rotation=75)
plt.tight_layout()
plt.show()
```

```python
rating = df.groupby(["type","rating"]).size().unstack().reset_index()
print(rating)
```

```python
m_rating =rating.reset_index().melt(id_vars='type', var_name='rating', value_name='count')
print(m_rating)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=m_rating, x='rating', y='count', hue='type')
plt.title("Movies vs TV Shows by Rating")
plt.xlabel("Rating Category")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
fig, axes = plt.subplots(2, 3, figsize=(18,10))  
axes = axes.flatten()


sns.barplot(data=data_90_to_21, x='release_year', y='count', hue='type', ax=axes[0])
axes[0].tick_params(axis='x', rotation=90)
axes[0].set_title('Number of Movies and TV Shows from 1991 to 2021')
axes[0].grid(True)

sns.barplot(data=above_avg_count, x='country', y='count', hue='type', ax=axes[1])
axes[1].tick_params(axis='x', rotation=90)
axes[1].set_title('Number of Movies and TV Shows by Country')
axes[1].set_ylabel("Number of Shows Produced")
axes[1].set_xlabel("Country")
axes[1].grid(True)

sns.barplot(x='genre', y='count', hue='type', data=df_melt, ax=axes[2])
axes[2].set_title('Comparison of Movies vs TV Shows by Genre')
axes[2].tick_params(axis='x', rotation=90)
axes[2].set_xlabel('Genre')
axes[2].set_ylabel('Number of Titles')
axes[2].grid(True)

sns.barplot(data=d_avg, x='director', y='count', color='#3b0a45', ax=axes[3])
axes[3].set_title('Top 10 Most Frequent Directors on Netflix')
axes[3].set_xlabel('Director')
axes[3].set_ylabel('Number of Shows/Movies')
axes[3].tick_params(axis='x', rotation=90)
axes[3].grid(True)

sns.barplot(data=m_rating, x='rating', y='count', hue='type', ax=axes[4])
axes[4].set_title("Movies vs TV Shows by Rating")
axes[4].set_xlabel("Rating Category")
axes[4].set_ylabel("Count")
axes[4].tick_params(axis='x', rotation=90)
axes[4].grid(True)


fig.delaxes(axes[5])

plt.tight_layout()
plt.show()
```
